```python
stp=stopwords.words("english")
```

```python
# By Using text pre-processing here are some of the steps:

# 1. Converting all the characters into Lowercase
# 2. Also remove/delete all the Unneccessary values/symbols or characters with the help of RegX
# 3. splitting all the Sentences/Data points into words by using word tokenizer
# 4. Later removing all the stp words (nothing but helping words)
# 5. And Stemming on each word by using porterstemming (any sort stemmer can be used)
```

```python
lf=[]
for sent in df["Reviews"]:
    l=[]
    ls=sent.lower()
    ls=re.sub("[^a-zA-Z ]","",ls)
    for word in word_tokenize(ls):
        if word in stp:
            pass
        else:
            stems=PorterStemmer().stem(word)
            l.append(stems)
    lf.append(" ".join(l))
```

```python
# The cleaned data is then added to create a DataFrame
```

```python
CLEAN_REVIEWS=pd.DataFrame({"cleand_reviewS":lf})
```

```python
os=[]
ps=[]
ls=[]
ss=[]
wl=[]
for y in df["Reviews"]:
    for word in word_tokenize(y):
        os.append(word)
        ps.append(PorterStemmer().stem(word))
        ls.append(LancasterStemmer().stem(word))
        ss.append(SnowballStemmer("english").stem(word))
        wl.append(WordNetLemmatizer().lemmatize(word))
```

```python
pd.DataFrame({"orignal_words":os,"poretr_s":ps,"lan_st":ls,"sno_stem":ss,"wl":wl})
```
